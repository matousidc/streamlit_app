```python
import pandas as pd
import altair as alt
import numpy as np
import os
from dotenv import load_dotenv
import requests
from sqlalchemy import create_engine, text
```

```python
load_dotenv(override=True)
connection_string = f"mysql+mysqlconnector://{os.getenv('USERNAME')}:{os.getenv('PASSWORD')}@{os.getenv('HOST')}/" \
                    f"{os.getenv('DATABASE')}?ssl_ca=/etc/ssl/cert.pem"
engine = create_engine(connection_string)
with engine.begin() as conn:
    # num_rows = df.to_sql('strava_dpnk', con=conn, if_exists='append', index=False)
    df_strava = pd.read_sql_query(text(f'SELECT * FROM strava_dpnk;'), con=conn)
```

```python
df = df_strava.drop_duplicates(subset=["name", "start_date"], ignore_index=True).sort_values(by="start_date", ascending=False).reset_index(drop=True)
df
```

```python
df = pd.read_pickle('activities_df.pkl')
print(df.shape)
df[-35:]
```

```python
# dff = df.iloc[df['dpnk' in df['name'].str.strip().str.lower()].index]
dff = df.iloc[df[df['name'].str.strip().str.lower() == 'dpnk'].index]
print(dff.shape)
# print(type(dff.loc[1, 'start_date']))
dff.head()
```

```python
df2 = df[(df['name'].str.strip().str.lower() == 'work commute') | (df['name'].str.strip().str.lower() == 'dpnk')].copy()
# df2.loc[:, 'name'] = df2.loc[:, "name"].str.lower().str.strip()
df2['start_date'] = pd.to_datetime(df2["start_date"])
print(df2.loc[1, "start_date"].hour)
df2 = df2[df2["start_date"].dt.hour < 12]
df2['date'] = pd.to_datetime(df2['start_date'].dt.strftime("2000-%m-%d"))
df2["year"] = df2["start_date"].map(lambda x: x.strftime("%Y"))
df2
```

```python
chart = alt.Chart(df2).mark_line(point=True).encode(x=alt.X('date:T', title="Date"), y=alt.Y('moving_time:Q', scale=alt.Scale(domain=[20, 27]), title="Moving time"), color=alt.Color("year", scale=alt.Scale(scheme='magma'), title="Year"), tooltip=['moving_time', alt.Tooltip('start_date:T')])
chart.properties(width=800, height=400)
```

```python
chart = alt.Chart(dff).mark_line(color='purple', point=True).encode(x=alt.X('start_date:T'), y=alt.Y('moving_time:Q', scale=alt.Scale(domain=[20, 26])), tooltip=['moving_time', alt.Tooltip('start_date:T')])
chart2 = alt.Chart(dff).mark_line(color='yellow').encode(x=alt.X('start_date:T'), y=alt.Y('elapsed_time:Q', scale=alt.Scale(domain=[20, 26], clamp=True)))
# color=alt.Color("start_date:T", scale=alt.Scale(scheme="goldred"))
(chart+ chart2).properties(width=800, height=400)
```

```python
cutoff = dff['moving_time'].mean()
print(cutoff)
chart = alt.Chart(dff).mark_point().encode(x=alt.X('start_date:T'), y=alt.Y('moving_time:Q', scale=alt.Scale(domain=[20, 26])), color=alt.condition(alt.datum.moving_time > cutoff, alt.value('green'), alt.value('red')), tooltip=['moving_time', alt.Tooltip('start_date:T')])# .add_params(alt.param(value=cutoff))
# chart = alt.Chart(dff).mark_point(color="purple").encode(x=alt.X('start_date:T'), y=alt.Y('moving_time:Q', scale=alt.Scale(domain=[20, 26])), tooltip=['moving_time', alt.Tooltip('start_date:T')])# .add_params(alt.param(value=cutoff))
# chart = alt.Chart(dff).mark_point().encode(x=alt.X('start_date:T'), y=alt.Y('moving_time:Q', scale=alt.Scale(domain=[20, 26])), color=alt.Color("moving_time", scale=alt.Scale(scheme='plasma')), tooltip=['moving_time', alt.Tooltip('start_date:T')])# .add_params(alt.param(value=cutoff))
reg = chart.transform_regression('start_date', 'moving_time').mark_line(color='red')
(chart + reg).properties(width=800, height=400)
```

```python
chart2 = alt.Chart(dff).mark_line(color="orange", point={"filled": True, "size": 50, "color": "orange"}).encode(x=alt.X('start_date:T'), y=alt.Y('moving_time:Q', scale=alt.Scale(domain=[20, 26])), tooltip=['moving_time', alt.Tooltip('start_date:T')])
alt.layer(chart2, reg).properties(width=800, height=400)
```
